¿Cuál fue la duración más frecuente de las películas en los años 90? Guarda una respuesta aproximada como un número entero llamada duration (usa 1990 como el año de inicio de la década).

Se considera que una película es corta si dura menos de 90 minutos. Cuenta el número de películas de acción cortas estrenadas en la década de 1990 y guarda este número entero como short_movie_count.

¡Siéntete libre de experimentar después de enviar el proyecto!

1. Filtrar solo por Movie

```python
peliculas= netflix_df[netflix_df['type']=='Movie']
```

2. Filtrar por año 1990 - 1999

```python
peliculas_90= peliculas[(peliculas['release_year']>=1990) & (peliculas['release_year']<=1999)]
```

```python
peliculas_90.head()
```

```python
peliculas_90.shape
```

##¿Cuál fue la duración más frecuente en los años 90?

```python
#Extraemos la moda
duration= peliculas_90['duration'].mode()[0]
print(int(duration))
```

Visualizar la distribución

```python
peliculas_90['duration'].hist(bins=20)
plt.xlabel("Duración (minutos)")
plt.ylabel("Cantidad de películas")
plt.title("Distribución de duración de películas (Años 90)")
plt.show()
```

##  Películas de acción cortas estrenadas en la década de 1990

```python
peliculas_accion=peliculas_90[(peliculas_90['genre']=='Action') & (peliculas_90['duration'] < 90)]
peliculas_accion
```

```python
short_movie_count= peliculas_accion.shape[0]
short_movie_count
```
